```python
import mysql.connector
import random
from datetime import datetime, timedelta

conn = mysql.connector.connect(
    host="localhost", user="root",
    password="Abhi@1445",  # change this
    database="tcg_market_analysis"
)
cur = conn.cursor()
random.seed(42)

# ── 1. MARKET DATA ────────────────────────────────────────
market = [
    ('Pune',   1460, 6100, 200, 6, 90),
    ('Mumbai',  2240, 3200, 120, 9, 150),
    ('Delhi NCR',1680,  2400, 150, 8, 120),
    ('Bangalore',1100, 2900, 140, 9, 80),
    ('Hyderabad',760,  1800, 130, 6, 80),
    ('Indore',  620,  1400, 110, 2, 45),
]
cur.execute("DELETE FROM market_data")
cur.executemany(
    "INSERT IGNORE INTO market_data VALUES (%s,%s,%s,%s,%s,%s)",
    market
)


# ── 2. DEMO TRACKER ───────────────────────────────────────
sales_people = ['Rahul S', 'Priya M', 'Vikram K', 'Sneha T']
cities       = ['Mumbai']*4 + ['Pune']*4 + ['Delhi NCR']*2
segments     = ['Small', 'Mid', 'Large', 'Enterprise']
outcomes     = ['Interested']*3 + ['Rejected']*5 + ['Follow_Up']*2

rejection_reasons = [
    'Post-sales module not ready',
    'Customer portal missing',
    'Exotel call quality issues',
    'Call drops frequently',
    'Price too high',
    'Already using Sell.Do',
    'Need more features before deciding',
]

demo_rows = []
start = datetime(2023, 8, 1)
for i in range(1, 151):
    city    = random.choice(cities)
    seg     = random.choice(segments)
    units   = {'Small':60, 'Mid':250, 'Large':900, 'Enterprise':3000}[seg]
    units  += random.randint(-20, 20)
    outcome = random.choice(outcomes)
    rr      = random.choice(rejection_reasons) if outcome == 'Rejected' else None
    ddate   = start + timedelta(days=random.randint(0, 300))
    fdate   = ddate + timedelta(days=7) if outcome == 'Follow_Up' else None
    demo_rows.append((
        i, random.choice(sales_people),
        f"Developer {i}", city, seg, units,
        ddate.strftime('%Y-%m-%d'),
        outcome, rr,
        fdate.strftime('%Y-%m-%d') if fdate else None
    ))

cur.executemany(
    "INSERT IGNORE INTO demo_tracker VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)",
    demo_rows
)

# ── 3. BUG TRACKER ────────────────────────────────────────
modules    = ['Exotel Integration']*5 + ['Pre-Sales Module']*3 + \
             ['AI Summary Engine']*3 + ['Marketing Module']*2
priorities = ['Critical']*2 + ['High']*3 + ['Medium']*4 + ['Low']*3
statuses   = ['Open']*4 + ['Resolved']*6

bug_rows = []
bstart = datetime(2023, 8, 1)
for i in range(1, 88):
    mod     = random.choice(modules)
    pri     = random.choice(priorities)
    stat    = random.choice(statuses)
    created = bstart + timedelta(days=random.randint(0, 200))
    res_days = random.randint(2, 20) if stat == 'Resolved' else None
    resolved = (created + timedelta(days=res_days)).strftime('%Y-%m-%d') \
               if res_days else None
    bug_rows.append((
        i, mod, pri, stat,
        created.strftime('%Y-%m-%d'), resolved
    ))

cur.executemany(
    "INSERT IGNORE INTO bug_tracker VALUES (%s,%s,%s,%s,%s,%s)",
    bug_rows
)

# ── 4. COMPETITOR FEATURES ────────────────────────────────
features = [
    (1,  'AI Call Summary',        1, 0, 0, 0),
    (2,  'AI Lead Pair Score',     1, 0, 0, 0),
    (3,  'Pre-Sales Module',       1, 1, 1, 1),
    (4,  'Marketing / Creative AI',1, 1, 0, 0),
    (5,  'Post-Sales Module',      0, 1, 1, 1),
    (6,  'Customer Portal',        0, 1, 0, 1),
    (7,  'Exotel Integration',     1, 1, 0, 0),
    (8,  'WhatsApp Messaging',     1, 1, 1, 0),
    (9,  '3D Renders / Visuals',   1, 0, 0, 0),
    (10, 'Inventory Management',   1, 1, 1, 1),
]
cur.executemany(
    "INSERT IGNORE INTO competitor_features VALUES (%s,%s,%s,%s,%s,%s)",
    features
)

conn.commit()
conn.close()
print("✅ Year 1 sample data ready!")
print("Market: 6 cities | Demos: 150 | Bugs: 87 | Features: 10")
```
